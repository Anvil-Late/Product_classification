# Custom functions used in the notebook

```python
def get_main_category(text):
    """
    Function that extract a product's main information
    ---
    Params :
    text : str from the product_category_tree variable
    """
    match = re.search('\["(.+?) >>', text)   
    if match:
        found = match.group(1)
        return(found)
    return("N/A")

def word_list_preprocessing(word_list):
    """
    Word preprocessor :
    - deletes stopwords
    - lowers string
    - stems words
    ---
    Params :
    word_list : liste of str
    """
    ps = PorterStemmer()
    
    output = [ps.stem(word.lower()) for word in word_list if word not in stopwords.words("english")]
    return(output)

def fully_tokenize(sentence):
    """
    Applies a tokenizer and calls the preprocessing function
    Params :
    sentence : str
    """
    tokenizer = nltk.RegexpTokenizer(r'\w+')
    temp_otp = tokenizer.tokenize(sentence)
    output = word_list_preprocessing(temp_otp)
    
    return(output)

def pad_to_size(array, x_shape, y_shape):
    """
    Applies padding to an array to reshape it to the desired
    shape
    ---
    Params :
    array : np.array to reshape
    x_shape : int, desired shape, ax nr 0
    y_shape : int, desired shape, ax nr 1
    
    """   
    output = np.zeros((x_shape, y_shape))
    if array is None:
        return(output)
    output[:array.shape[0],:array.shape[1]] = array
    
    return(output)

def silhouette_dataviz(X, n_cluster_range, two_axes = True):
    """
    Plots the silhouette score for each sample of a dataset for several K values 
    of a K-means clustering. Can also provides a dataviz on a TSNE manifold
        
    ---
    Params :
    X : Matrix or dataframe that will be K-means clustered
    n_cluster_range : range of K values to test
    two_axes : set to true if you want the TSNE dataviz. 
               the TSNE matrix must be made beforehand
    """
    
    # Si on choisit de plot la silhouette ET la dataviz sur T-SNE 
    if two_axes: 
        for n_clusters in n_cluster_range:
            # Création d'un subplot avec 1 ligne & 2 colonnes 
            fig, (ax1, ax2) = plt.subplots(1, 2)
            fig.set_size_inches(18, 7)

            # 1er subplot : silhouette
            ax1.set_xlim([-1, 1])
            # le (n_clusters+1)*10 laisse un espace blanc entre les silhouettes 
            # de chaque cluster pour les distinguer plus facilement
            ax1.set_ylim([0, X.shape[0] + (n_clusters + 1) * 10])

            # Initialisation du cluster avec reproducibilité 
            clusterer = KMeans(n_clusters = n_clusters, random_state = 42)
            cluster_labels = clusterer.fit_predict(X)

            # Calcul du score de silhouette moyen
            silhouette_avg = silhouette_score(X, cluster_labels)
            print("For n_clusters =", n_clusters,
                  "the silhouette score is :", silhouette_avg)

            # Calcul du score de silhouette pour chaque échantillon
            sample_silhouette_values = silhouette_samples(X, cluster_labels)
            
            # Définition du y de début de tracé du subplot
            y_lower = 10
            for i in range(n_clusters):
                # Groupage des scores de silhouette de chaque échantillon du cluster i
                # puis tri des valeurs 
                ith_cluster_silhouette_values = \
                    sample_silhouette_values[cluster_labels == i]

                ith_cluster_silhouette_values.sort()

                size_cluster_i = ith_cluster_silhouette_values.shape[0]
                y_upper = y_lower + size_cluster_i

                color = cm.nipy_spectral(float(i) / n_clusters)
                
                ax1.fill_betweenx(np.arange(y_lower, y_upper),
                                  0, ith_cluster_silhouette_values,
                                  facecolor=color, edgecolor=color, alpha=0.7)

                # Labelisation du silhouette plot
                ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

                # Calcul du y_lower du prochain plot 
                y_lower = y_upper + 10  

            ax1.set_title("Silhouette score by clustering")
            ax1.set_xlabel("Silhouette score")
            ax1.set_ylabel("Cluster number")

            # Tracé du score de silhouette moyen en rouge pointillé 
            ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

            ax1.set_yticks([])  # Reset les y labels / ticks
            ax1.set_xticks([-1, -0.8, -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6, 0.8, 1])

            # 2e subplot : visualisation sur T-SNE

            for i in range(n_clusters):
                color = cm.nipy_spectral(float(i) / n_clusters)
                ax2.scatter(tsne_data[cluster_labels == i , 0] , tsne_data[cluster_labels == i, 1] , color = color, label = i)


            ax2.set_title("Dataviz on TSNE")
            ax2.set_xlabel("")
            ax2.set_ylabel("")
            ax2.legend()
            plt.suptitle(("Silhouette analysis for a K-means clustering with "
                          "n_clusters = %d" % n_clusters),
                         fontsize=14, fontweight='bold')
            
    # Si on ne veut que le silhouette plot sans le T-SNE :
    # même code sans la fin :
    else:
        for n_clusters in n_cluster_range:
            
            fig, ax1 = plt.subplots(1, 1)
            fig.set_size_inches(18, 7)

           
            ax1.set_xlim([-0.2, 1])
            
            ax1.set_ylim([0, X.shape[0] + (n_clusters + 1) * 10])

           
            clusterer = KMeans(n_clusters = n_clusters, random_state = 42)
            cluster_labels = clusterer.fit_predict(X)

            
            silhouette_avg = silhouette_score(X, cluster_labels)
            print("For n_clusters =", n_clusters,
                  "the silhouette score is :", silhouette_avg)

           
            sample_silhouette_values = silhouette_samples(X, cluster_labels)

            y_lower = 10
            for i in range(n_clusters):
              
                ith_cluster_silhouette_values = \
                    sample_silhouette_values[cluster_labels == i]

                ith_cluster_silhouette_values.sort()

                size_cluster_i = ith_cluster_silhouette_values.shape[0]
                y_upper = y_lower + size_cluster_i

                color = cm.nipy_spectral(float(i) / n_clusters)
              
                ax1.fill_betweenx(np.arange(y_lower, y_upper),
                                  0, ith_cluster_silhouette_values,
                                  facecolor=color, edgecolor=color, alpha=0.7)

               
                ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

               
                y_lower = y_upper + 10 

            ax1.set_title("Silhouette score by clustering")
            ax1.set_xlabel("Silhouette score")
            ax1.set_ylabel("Cluster number")

           
            ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

            ax1.set_yticks([])
            ax1.set_xticks([-0.2, 0, 0.2, 0.4, 0.6, 0.8, 1])
            
            plt.suptitle(("Silhouette analysis for a K-means clustering with "
                          "n_clusters = %d" % n_clusters),
                         fontsize=14, fontweight='bold') 

    plt.show()
```

We'll use this to create bags of words

## Bag of words

We'll create a bag of word with the following loop :

```python
wordfreq = {}
for sentence in data["description"]:
    tokenized_sentence = fully_tokenize(sentence)
    for token in tokenized_sentence :
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1
```

We end up with a dictionnary that contains more than 5000 tokens : 

```python
len(wordfreq.keys())
```

However, most of the tokens appear only once :

```python
(q1, q2, q3, q4) = np.quantile(list(wordfreq.values()), [0.25, 0.5, 0.75, 1])
for (qtl, qv) in zip([0.25, 0.5, 0.75, 1], (q1, q2, q3, q4)):
    print("{}% of tokens appear {} times or less".format(qtl * 100, qv))
```

Left as it is, we would end up with a bag of words matrix that has 1050 rows and 5077 columns that are mainly empty. This size is acceptable given my RAM but for bigger datasets, we would have had to filter out tokens with low occurence.

We can now convert our dictionary to a bag of word matrix

```python
tf = CountVectorizer(tokenizer = fully_tokenize)
tf_matrix = tf.fit_transform(data["description"])
```

Let's see what our matrix is like when represented into the following plot :

```python
fig, ax = plt.subplots(figsize = (16, 8))
ax.imshow(tf_matrix.toarray(), cmap = plt.cm.Reds, interpolation = "nearest", aspect = "auto")
plt.suptitle("Visual representation of the bag of words matrix \n 0 values are in beige \n 1 values are in red")
plt.xlabel("Column number")
plt.ylabel("Row number");
```

We'll see what we get when we apply a TF-IDF vectorization rather than a basic bag of words.

## TF-IDF

```python
tfidf = TfidfVectorizer(tokenizer = fully_tokenize)
tfidf_matrix = tfidf.fit_transform(data["description"])
```

```python
fig, ax = plt.subplots(figsize = (16, 8))
ax.imshow(tfidf_matrix.toarray(), cmap = plt.cm.Reds, interpolation = "nearest", aspect = "auto")
plt.suptitle("Visual representation of the TF-IDF matrix \n 0 values are in beige \n > 0 values are in red")
plt.xlabel("Column number")
plt.ylabel("Row number");
```

The matrix appears to be less sparse, but still sparse enough to make a dimension reduction advisable

### Dimension reduction

```python
mm_tfidf = MinMaxScaler().fit_transform(tfidf_matrix.toarray())
```

```python
pca2 = PCA(0.8, svd_solver = "full", random_state = 42)
pca2.fit(mm_tfidf)
pca2.n_components_
```

With only 318 PCs, from the 5077 initial variables, we keep 80% of the variance. Let's see how that changes given a different number of PCs :

```python
pca2_cumsum = pca2.explained_variance_ratio_.cumsum()

fig, ax = plt.subplots(figsize = (18, 8))
ax.bar(range(pca2_cumsum.shape[0]), pca2_cumsum, alpha = 0.7);
ax.set_xlabel("Number of principal components", fontsize = 14);
ax.set_ylabel("Explained variance", fontsize = 14);
ax.set_xticks(range(0, pca2_cumsum.shape[0] + 1, 5));

for yvalue in [0.1 * i for i in range(1, 9)]:
    ax.axhline(yvalue, alpha = 0.2, color = "green")
plt.xticks(rotation = 45, rotation_mode = "anchor", ha = "right");
plt.suptitle("Explained variance by the PCA based on the number of PCs",
            fontsize = 18, fontweight = "bold");
```

### Hyperparameter tuning

Usually we recommend using 1 principal component for every 100 rows of a dataset. If you speak french, [here's a very interesting video that gives more detail on this principle (be careful, it is math heavy)](https://www.youtube.com/watch?v=WtJl5ez0hfY&list=PLtzmb84AoqRTl0m1b82gVLcGU38miqdrC&index=14).

Here, since we have 1050 observations, we'll use 11 principal components

```python
pca = PCA(11, random_state = 42)
pca.fit(mm_tfidf)
pca_tfidf = pca.transform(mm_tfidf)
```

```python
inertias = []

K = range(2, 21)

for k in K:

    # Création du modèle
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(pca_tfidf)
    
    # Calcul de l'intertie
    inertias.append(kmeanModel.inertia_)


fig, ax = plt.subplots(figsize = (18, 8))
ax.plot(K, inertias, 'bx-')
ax.set_xlabel('K values', fontsize = 14)
ax.set_ylabel('Inertia', fontsize = 14)
ax.set_xticks([i for i in range(21)])
plt.suptitle('Elbow Method by inertia', fontsize = 18)
plt.show()
```

```python
start_clock = time.time()
silhouette_dataviz(pca_tfidf, range(2, 21), two_axes = True)
print("Silhouette analysis done in {}s".format(round(time.time() - start_clock)))
```

The elbow analysis suggests a K value of 13 or 14, while the silhouette analysis gives a maximal score for K = 12 and K = 14. 

We'll go with K = 14, not only because it is suggested by both analyses, but because 14 is twice the original number of categories, which intuitively makes it more likely to get a clustering similar to the original classification.

```python
for K in [14]:
    kmeans = KMeans(n_clusters = K, random_state = 42)
    kmeans.fit(pca_tfidf)
    kmeans_labels = kmeans.predict(pca_tfidf)

    km_labs_distribution = pd.DataFrame(np.array(np.unique(kmeans_labels, return_counts = True)).T)
    km_labs_distribution.rename(columns = {0 : "cluster", 1 : "count"}, inplace = True)
    km_labs_distribution["proportion"] = km_labs_distribution["count"].apply(lambda X : round(100 * X / km_labs_distribution["count"].sum()))

    fig, ax = plt.subplots(figsize = (18, 8))
    ax.bar(km_labs_distribution["cluster"], km_labs_distribution["proportion"])
    #ax.set_xticks(dbs_labs_distribution["cluster"]);
    ax.set_xlabel("Cluster", fontsize = 14)
    ax.set_ylabel("% of observation in the cluster", fontsize = 14)
    plt.suptitle("K-means clustering distribution, K = {}".format(K),
                 fontsize = 18)
```

We appear to have 7 major clusters and 7 minor ones, which is intuitively good news given our original objective.

## Loading images

For every product, we'll use the *image* column to import the associated image from our computer. We'll store them into a dictionary :

```python
img_dict = {}
```

```python
start_clock = time.time()
for item_id in data["uniq_id"]:
    item_image_id = data.loc[data["uniq_id"] == item_id, "image"].iloc[0]
    item_image_path = "./data/Flipkart/Images/" + item_image_id
    #item_image = Image.open(item_image_path)
    item_image = cv2.imread(item_image_path,0)
    item_image = cv2.resize(item_image, (400, 400))
    img_dict[item_id] = item_image
print("Images loaded in {}s".format(round(time.time() - start_clock)))
```

We'll use the SIFT algorithm to extract features from images. Its patent has expired in [july 2020](https://www.haikutechcenter.com/2020/07/patent-expired-on-sift.html), after what it has been implemented in the *cv2* package for versions 4.4 and above. For those of you that want to run this notebook and have an older version, the code chunk below handles both versions with a try / except call :

```python
try : # if cv2 version < 4.4
    sift = cv2.xfeatures2d.SIFT_create()
except : # if cv2 version >= 4.4
    sift = cv2.SIFT_create()
```

We create a SIFT descriptor for every image :

```python
desc_list = {}
i = 0
start_clock = time.time()
for key in img_dict.keys():
    i += 1
    if i % 100 == 0 :
        print("Processing image nr {}...".format(i))
    temp_array = np.array(img_dict[key])
    kp, desc = sift.detectAndCompute(temp_array, None)
    desc_list[key] = desc
print("SIFT descriptors created in {}s".format(round(time.time() - start_clock)))
```

## Padding

Here, we'll try to make all SIFT descriptors' shapes the same

```python
max_x = 0
max_y = 0
```

```python
for key in desc_list.keys():
    try :
        desc_x = desc_list[key].shape[0]
        desc_y = desc_list[key].shape[1]
    except AttributeError:
        print("Error with key nr {}".format(key))
        
    
    if desc_x > max_x:
        max_x = desc_x
    if desc_y > max_y:
        max_y = desc_y
        
print("max_x : {}".format(max_x))
print("max_y : {}".format(max_y))
```

### Side study : key error

We previously got an error for key nr 9743e6d1c9f3180483825fe4d2e97836. Let's see what the problem is

```python
error_key = "9743e6d1c9f3180483825fe4d2e97836"
```

```python
desc_list[error_key]
```

```python
img_dict[error_key]
```

The error comes from the fact that the picture is completely white after compression. Therefore, the SIFT algorithm finds no features to describe and returns an empty array.

The original image is a white macbook on a white background.

### Back to padding

Now that we know the maximal size that our SIFT descriptors can take, we can paddy our matrices with our custom function *pad_to_size* which works like this :

```python
test_array = np.array([[1, 1], [2, 2]])
print(test_array)
```

```python
pad_to_size(test_array, 5, 3)
```

Our function can also handle errors like the one from key number 9743e6d1c9f3180483825fe4d2e97836 by returning a sparse matrix :

```python
pad_to_size(None, 5, 3)
```

We apply this function to our whole dictionary :

```python
for key in desc_list.keys():
    desc_list[key] = pad_to_size(desc_list[key], max_x, max_y)
```

## Model building

We will now build our model with the following steps :

1. Defining X and y data
2. Encoding X et y
3. Model building
4. Model training
5. Model testing

here, our X and y data are respectively our "messages" vector and y is our main_category variable.

```python
X = messages
y = data["main_category"].to_numpy()
```

As for the image reconition model, we need to pre-process our response variable so that it has 7 dimensions. To do that, we create an encoding and decoding function :

```python
le = preprocessing.LabelEncoder()
le.fit(y)
```

```python
def encode(le, labels):
    enc = le.transform(labels)
    return keras.utils.to_categorical(enc)

def decode(le, one_hot):
    dec = np.argmax(one_hot, axis = 1)
    return(le.inverse_transform(dec))
```

Here's how our encoding function works :

```python
test = encode(le, ["Watches", "Home Furnishing", "Home Furnishing", "Baby Care"])
test
```

And here's how the decoding fuction works :

```python
untest = decode(le, test)
untest
```

We encode y. I personally like to have uniformity in my variable names so I will also rename X as X_enc even though I don't change it.

After that, we split the data into a training set and a testing set.

```python
X_enc = X
y_enc = encode(le, y)

X_train, X_test, y_train, y_test = train_test_split(X_enc, y_enc, test_size=0.3, random_state=42)
```

Our resulting datasets have the following shape :

```python
for subelement in (X_train, X_test, y_train, y_test):
    print(subelement.shape)
```

We now create our neural network that will contain :

* An input layer
* An embedding layer, which will be the universal sentence encoder
* Two dense layers that will result in a 7-class classification

```python
def UniversalSentenceEncoding(x):
    return embed(tf.squeeze(tf.cast(x, tf.string)))
```

```python
input_layer = Input(shape = (1,), dtype = tf.string)
embedding_layer = Lambda(UniversalSentenceEncoding, output_shape = (512,))(input_layer)
dense_layer = Dense(256, activation = "relu")(embedding_layer)
prediction_layer = Dense(7, activation = "softmax")(dense_layer)

nlp_model = Model(inputs = [input_layer], outputs = prediction_layer)
```

We end up with the following model :

```python
nlp_model.summary()
```

Our model has very few parameters to train because the majority of the work is done beforehand by the encoder. What we did is therefore transfer learning.

We can now compile and train our model.

```python
nlp_model.compile(loss='categorical_crossentropy', optimizer = "adam", metrics=["accuracy"])
callback = EarlyStopping(monitor='val_loss', patience=3, min_delta = 0.008, restore_best_weights = True)
```

```python
start_clock = time.time()
with tf.Session() as session:
    KB.set_session(session)
    session.run(tf.global_variables_initializer())
    session.run(tf.tables_initializer())
    training_session = nlp_model.fit(X_train, y_train, epochs = 50, batch_size = 50, verbose=0, 
                                     validation_split = 0.3, callbacks = [callback])
    nlp_model.save_weights("./data/nlp_model.h5")
print("Model trained in {}s".format(round(time.time() - start_clock)))
```

# Conclusion

We made basic and advanced models in both image recognition and NLP. Let's compare the results :

```python
methods = ["Basic NLP", "Basic \n image \n recognition"]
method_silhouette = [0.5651547479274082, 0.5243233610956617]
method_ari = [0.29233258710356447, 0.05216561117769667]

summary_df = pd.DataFrame()
```

```python
summary_df["methods"] = methods
summary_df["Silhouette score"] = method_silhouette
summary_df["ARI score"] = method_ari
```

```python
fig, ax = plt.subplots(figsize = (18, 8))
summary_df.plot(kind = "bar", ax=ax)
plt.xticks(range(0,len(summary_df["methods"])), summary_df["methods"]);
plt.setp(ax.get_xticklabels(), fontsize = 12, rotation=45, ha="right", rotation_mode="anchor");
plt.suptitle("Basic unsupervized model performances", fontsize = 18, fontweight = "bold");
ax.set_ylabel("Score", fontsize = 16);
ax.set_xlabel("Method", fontsize = 16);
plt.show()

fig, ax = plt.subplots(figsize = (9, 8))
ax.bar(supervized_model_score.keys(), supervized_model_score.values(), color = "green", width = 0.4)
for idx, score in enumerate(list(supervized_model_score.values())):
    ax.text(idx - 0.05, score + 0.02, round(score, 2), fontweight = "bold", fontsize = 12, color = "k")
plt.setp(ax.get_xticklabels(), fontsize = 12, rotation=45, ha="right", rotation_mode="anchor");
plt.suptitle("Deep Learning classifier performances", fontsize = 18, fontweight = "bold");
ax.set_ylabel("Accuracy", fontsize = 16);
ax.set_xlabel("Method", fontsize = 16);
ax.set_ylim(0, 1);
```

While unsupervized models end up having good silhouette scores, they struggle to match the original classification. They can be used to discover new categories that weren't identified, but it would be best to avoid using them as actual classifiers, which would also force us to label all of the clusters anyways.

**Therefore, the best models would be the supervized neural networks. By labeling a small subset of our catalog, we can train very efficient classifiers that can automate category labeling with a high accuracy. If we lack resources, the best one would be the NLP model that performs better and is trained very fast, which means it is more scalable.**

**If we do have the time and resources to do so, the best method would be to use both neural networks, with the NLP model as the main classifier. When the model becomes unsure about a prediction, we can call the image recognition model to check if the doubt can be lifted, and end up with an ensemble model that can be even more accurate.**
